# Notebook 01 — Data Collection

Downloads all commodity price and weather data required for the analysis.

**Improvement over original paper:**
- Uses `yfinance` for automated, reproducible data download
- Multi-station production-weighted weather composites
- Data extended to 2024 (original: up to Nov 2019)


```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from src.data.fetch_commodity_data import fetch_all_commodities, descriptive_stats
from src.data.fetch_weather_data import COMMODITY_STATIONS
```

```python
# Fetch all commodity prices (1973–2024)
prices = fetch_all_commodities(
    start='1973-01-01',
    end='2024-12-31',
    save=True
)
print(f'Shape: {prices.shape}')
print(f'Date range: {prices.index.min()} → {prices.index.max()}')
prices.tail()
```

```python
# Descriptive statistics — replicates Table 4.1 from original paper
stats = descriptive_stats(prices)
print('Descriptive Statistics — Raw Prices')
display(stats)
```

```python
# Show weather station configuration (multi-station improvement)
print('Weather Station Configuration (multi-station composites)')
print('='*60)
for commodity, stations in COMMODITY_STATIONS.items():
    countries = [f"{s['country']} ({s['weight']:.0%})" for s in stations]
    print(f'  {commodity:<10}: {", ".join(countries)}')
```
